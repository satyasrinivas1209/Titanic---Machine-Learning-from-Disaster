```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame (replace with your real dataset)
data = {
    'Age': [22, 38, 26, 35, None, 54, 2, 27, 14, 4],
    'Fare': [7.25, 71.83, 7.92, 53.10, 8.05, 51.86, 21.07, 11.13, 30.07, 16.70]
}
df = pd.DataFrame(data)

# Drop NaN values before plotting
df['Age'].dropna().hist(bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

df['Fare'].dropna().hist(bins=20, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame (replace with your actual dataset)
data = {
    'Pclass': [1, 2, 3, 1, 3, 2, 1, 3],
    'Age': [22, 38, 26, 35, None, 54, 2, 27],
    'Survived': [0, 1, 1, 1, 0, 0, 1, 0],
    'Fare': [71.83, 7.92, 53.10, 8.05, 51.86, 21.07, 11.13, 30.07]
}
df = pd.DataFrame(data)

# Boxplot 1 - Age by Passenger Class
plt.figure(figsize=(6, 4))
sns.boxplot(x="Pclass", y="Age", data=df)
plt.title("Age by Passenger Class")
plt.show()

# Boxplot 2 - Fare by Survival
plt.figure(figsize=(6, 4))
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare by Survival")
plt.show()

```

```python
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Age vs Fare by Survival")
plt.show()
```
